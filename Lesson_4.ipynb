```python
# 1. Загрузите модуль pyplot библиотеки matplotlib с псевдонимом plt, а также библиотеку numpy с псевдонимом np.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pylab import rcParams
```

```python
# Примените магическую функцию %matplotlib inline для отображения графиков в Jupyter Notebook и настройки 
# конфигурации ноутбука со значением 'svg' для более четкого отображения графиков.

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
rcParams['figure.figsize'] = 6, 4.5
```

```python
# Создайте список под названием x с числами 1, 2, 3, 4, 5, 6, 7 и список y с числами 3.5, 3.8, 4.2, 4.5, 5, 5.5, 7.

x=[1, 2, 3, 4, 5, 6, 7]
y=[3.5, 3.8, 4.2, 4.5, 5, 5.5, 7.]
print(x)
print(y)
```

```python
# С помощью функции plot постройте график, соединяющий линиями точки с горизонтальными координатами из списка x 
# и вертикальными - из списка y

plt.plot(x, y)
plt.show()
```

```python
# Затем в следующей ячейке постройте диаграмму рассеяния (другие названия - диаграмма разброса, scatter plot).

plt.scatter(x, y)
plt.show
```

```python
# 2. С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.

t = np.linspace(0, 10 , 51)
t
```

```python
# Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.

f = np.cos(t)
f
```

```python
# Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. 
# Линия графика должна быть зеленого цвета.

plt.plot(t, f, color = 'green')
plt.show()
```

```python
# Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' 
# и для вертикальной - 'Значения f'.

plt.plot(t, f, color = 'green')

title_dict = {'fontsize': '16', 'color': 'brown', 'family': 'Calibri', 'fontweight': 'bold'}
label_dict = {'fontsize': '14', 'color': 'Blue', 'family': 'Calibri'}

plt.title('График f(t)', fontdict = title_dict)
plt.xlabel('Значение t', fontdict = label_dict)
plt.ylabel('Значение f', fontdict = label_dict)

# Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.grid(color = 'ghostwhite')

plt.show()
```

```python
# 3. С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

x = np.linspace(-3, 3, 51)
x

# Создайте массивы y1, y2, y3, y4 по следующим формулам:
# y1 = x**2
# y2 = 2 * x + 0.5
# y3 = -3 * x - 1.5
# y4 = sin(x)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

print(y1, y2, y3, y4, sep = '\n\n')
```

```python
# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig 
# и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из 
# двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем 
# графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.
# Дайте название графикам: 'График y1', 'График y2' и т.д.

fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')

# Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
ax1.set_xlim([-5, 5])
ax1.set_ylim([0, 9])
ax1.grid(color = 'ghostwhite')

ax2.plot(x, y2)
ax2.set_title('График y2')
ax2.grid(color = 'ghostwhite')

ax3.plot(x, y3)
ax3.set_title('График y3')
ax3.grid(color = 'ghostwhite')

ax4.plot(x, y4)
ax4.set_title('График y4')
ax4.grid(color = 'ghostwhite')

#Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
fig.set_size_inches(8, 6)

#Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.
plt.subplots_adjust(hspace = 0.3, wspace = 0.3)
```

```python
# 4. В этом задании мы будем работать с датасетом, в котором приведены данные по мошенничеству с кредитными данными:
# Credit Card Fraud Detection (информация об авторах: Andrea Dal Pozzolo, Olivier Caelen, Reid A. Johnson and 
# Gianluca Bontempi. Calibrating Probability with Undersampling for Unbalanced Classification. In Symposium on 
# Computational Intelligence and Data Mining (CIDM), IEEE, 2015).
# Ознакомьтесь с описанием и скачайте датасет creditcard.csv с сайта Kaggle.com по ссылке: Credit Card Fraud Detection
# Данный датасет является примером несбалансированных данных, так как мошеннические операции с картами встречаются реже обычных.
# Импортруйте библиотеку Pandas, а также используйте для графиков стиль “fivethirtyeight”.

#используйте для графиков стиль fivethirtyeight. 
plt.style.use('fivethirtyeight')

#загрузить датасет 
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# Посчитайте с помощью метода value_counts количество наблюдений для каждого значения целевой переменной Class 
# и примените к полученным данным метод plot, чтобы построить столбчатую диаграмму. 
class_value = df['Class'].value_counts()

df_class_info = pd.Series(class_value)
df_class_info.plot(kind = 'bar')
plt.show()
```

```python
# Постройте такую же диаграмму, используя логарифмический масштаб.
df_class_info = pd.Series(class_value)
df_class_info.plot(kind = 'bar', logy = True)
plt.show()
```

```python
# На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций 
# (Class равен 1) и другую - для обычных (Class равен 0). Подберите значение аргумента density так, чтобы 
# по вертикали графика было расположено не число наблюдений, а плотность распределения. Число бинов должно 
# равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными 
# и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных 
# транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.

df.value_counts()

class0 = df.loc[df['Class'] == 0, ['V1']]
class1 = df.loc[df['Class'] == 1, ['V1']]

plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend(frameon = False)
plt.xlabel('V1')
plt.axis([-32, 3, 0, 0.2])
plt.show()
```
